R里的`rank`返回的向量的下标的值是输入向量的下标的值在输入向量中的排序

R里的`order`返回的向量的下标的值是输入向量排序后的向量的值在输入向量的下标

```r
v <- rnorm(10)
n <- length(v)

v
```

```r
v[order(v)] == sort(v)
```

```r
(1:n)[rank(v)] == rank(v)
```

## 下标复合

假如一个从1取到n的排列是指从1,2,...,n一一映到那个排列的变换。则一个右式复合是指两个这样的变换的合成。

例：

$$
g: \\
1 \to 1 \\
2 \to 3 \\ 
3 \to 2 \\
f: \\
1 \to 2 \\
2 \to 1 \\
3 \to 3 \\
f \circ g : \\
1 \to 1 \to 2  \\
2 \to 3 \to 3 \\
3 \to 2 \to 1
$$

R中可以以$g$的向量放到f的向量的subscript里来表示

```r
g <- c(1,3,2)
f <- c(2,1,3)
fg <- f[g]
fg
```

可以看出这样设计还是很符合复合原则的

## rank迭代

rank不断迭代虽然很快进入周期，但本身没什么特别的规律。（尤其有限性，任意状态开始显然迟早陷入一个周期中。）

```r
rank.iter <- function(arr, n){
    rl <- list()
    for(i in 1:n){
        rl[[i]] <- rank(arr)
        arr <- arr[rank(arr)]
    }
    return(rl)
}
```

```r
rank.iter(v,20)
```
