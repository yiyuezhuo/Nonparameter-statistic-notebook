# 置信区间与假设检验

某个总体关于某个参数的一个$1-a$置信水平的置信区间，是指该随机区间以$1-a$概率包含参数。置信区间由样本计算出来，样本是随机的，
故置信区间是随机的。

从假设检验中可以反转出置信区间。我们可以这么做，首先我们指定一类假设检验，这类假设检验每个检验不同的零假设，零假设就是总体参数等于某个数，
每个不同的假设里也使用不同的统计量，通常其中有常数取成了零假设中钦定的那个数。然后我们固定样本值，看看该样本下哪些参数对应的假设检验没有拒绝，
把这些参数做成一个集合，该从样本到集合的对应规则就是一个$1-a$置信水平（$a$的那类假设检验中的显著性水平）的置信区间。

## 置信区间的一些特别例子

置信区间试图估计的那个参数一般是不知道的，然后我们的置信区间可以对所有参数取法，均做到$1-a$置信水平又不依赖参数。
然而如果这个参数的任意性受到了削弱，我们就可以通过一些特例把它看得更清楚。

如果我们实际知道那个参数就是取$0$，我们可以设固定区间为$[-0.001,0.001]$，它以100%置信水平包含参数。100%置信水平在任取情况也可以做到，
取整个值域即可。如果我们知道那个参数只能取$0,1,\ldots,n$，
则我们可以类似的设$[-0.001,0.001] \cup [0.999,1.001] \cup \dots \cup [n-0.001,n+0.001]$来做到100%置信区间。
当然它们都可以通过包含整个值域来做到100%置信水平。

既然做到某一置信水平并不是什么困难的事，于是我们要求置信区间同时做一些其他优化，就像什么最优功效检验一样，它们不一定存在。

## 置信区间式对象的抽象描述

我们实际在做这么一件事，对于一个分布类，对于其中所有分布和所有样本规模，我们的样本到集合的函数，的随机集合包含从分布到集合元素的函数这一事件的概率都等于某值。

置信区间这么特例上，分布类中的每个分布起码要有一个地位类似的待估计的参数，那个从分布到集合元素的函数就从分布把这个参数映射出来。
这些分布给出类似形状的样本，使得样本到集合函数总可以对应一个集合。容易看出分布类中若包含的分布减少，
我们之前满足要求的分布到集合函数仍保持相同的事件概率，同时一些在新集合上可以保持，在去掉的分布上不能保持的函数也可以作为新的，符合定义的函数。
在最极端的只包含一个特定分布的情况中，由于参数只有一个，我们完全可以仿照上面的方法搞一个固定区间。而这种函数显然在更一般的分布类中是无效的。
(样本规模是一个地位特殊的参数，通常我们通过那个集合函数中的原象包含不同样本规模的样本。虽然严格说，分布必须结合样本规模才是完全的)

置信区间本身的一个有名特例是包含所有正态分布(即所谓期望和方差都未知的意思)的分布类。这个上面可以利用标准化统计量搞出置信区间，比如

$$
\frac{n(\bar{X} - EX)}{s_X} \sim t(n-1) \\
P(t(n-1,\alpha/2) < \frac{n(\bar{X} - EX)}{s_X} < t(n-1,1-\alpha/2)) = 1 - \alpha
$$
