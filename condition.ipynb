# 条件概率系统

不基于测度论的条件概率系统是一个非常混乱，拥有蹩脚记号的丑陋体系。

## 多维随机变量的样本空间解释

$$
X + Y > k
$$

像$X > k$这样的写法我们知道表示的是$\{s \, \vert \, s \in \Omega \land X(s) > k \} \subset \Omega$这个符合条件的样本空间的子集。
那么$X+Y$呢？

试想$\{ 0,1, \ldots , 9 \} \times \{ 0,1, \ldots, 9 \} =\Omega$，令$X(s) = X((a,b)) = a,Y(s) = Y((a,b)) = b$。
$P(s) = \frac{1}{100}$。我们写出$X < 1.5$时表达的是$\{ (0,0),(0,1), \ldots, (0,9) ,(1,0), \ldots (1,9)\} $ 这样的包含20个样本点的集合。
$X + Y < 1.5$则表达是$\{ (0,0),(0,1),(1,0) \}$这样有三个元素的集合。

当我们令$Z = X + Y$时，有$Z(s) = X(s) + Y(s) = Z((a,b)) = a + b$。显然$Z < 1.5$与$X+Y < 1.5$表示的是同一个集合。

有时会试图将$X | Y = y$之类的写法看成单独的成分，显然这似乎并没有什么意义，因为$X$如何映射样本空间的自然定义就是它原来的定义，
怎么“条件”都没有任何意义。

上述成分和测度函数$P$或期望算子$E$上组合产生了不同的定义。
$$
P(X = x | Y = y) = \frac{P(X = x \cap Y = y)}{P(Y = y)}
$$
当$P(Y = y) \neq 0$时成立，若不幸成立（连续中的常见情况），则必须借助测度论才能精确定义。

在离散情况可以特别定义条件期望为
$$
E(X | Y = y) = \sum_{p \in \Omega} P(\{ p\}  | Y = y) X(p)
$$

由于一般我们都是用随机变量的比较运算来反推出可测集，而不是直接写出。所以上述混合写法显得很有启发性。
这诱使我们写出在“限制在”（条件化到）某个可测集上的诱导测度，

$$
P(S_1 | S_2) = \frac{P(S_1 \cap S_2)}{P(S_2)}
$$

在$P(S_2) \neq 0$下成立，显然它满足只在$S_2$外取值时为$0$。在$S_2$内部则统一放大测度值。

如随机变量$X$是简单函数（只映射到有限个实数上，如有限值离散随机变量），$X(s_i) = x_i \quad \forall p_i \in S_i \quad \cup_i S_i = \Omega$，可以定义期望为

$$
\sum_{i = 1}^n P(S_i) X(s_i)
$$

n是所分类数。

于是在上述条件都成立这种极其简化的情况中，可以定义条件期望为条件诱导后($Y = k,Y < k$之类的条件显然可以作为$S_2$来处理)。
在诱导后的测度上算普通期望。

$$
E(X | S_2) = \sum_{i=1}^n P_2(S_i) X(s_i) \\
P_2(S_i) = \frac{P(S_i \cap S_2)}{P(S_2)}
$$

容易看出上述简单函数的期望其实就是简单函数$X(s_i)$的积分。由测度论可以从简单函数建立一般函数的积分。这样就可以定义出一般的随机变量的期望，
再将条件诱导期望一般化就可以得到一般的条件期望了。
